```python
#1) class <class_name>:
class subway_frog:
    __frogs_count = 0 # Class attribute

    def run(self):
        self.__state = "running"
        
    def jump(self):
        self.__state = "jumping"
        
    def roll(self):
        self.__state = "rolling"
        
    def hit(self):
        self.__state = "hit"
        if self.power == "no":
            print("You are caught!!! Game over ")
        else:
            print("As you have power, you are not caught!!! Continue")
            self.power = "no"

    def run_over_gift(self):
        self.__get_power("power1")
            
    def __get_power(self, power):
        self.power = power

    # 2) Constructor: __init__
    def __init__(self, color="red", name=" "):
        self.__state = "idle" # Private attribute
        self.color = color # Public attribute
        self.name = name
        self.__get_power("no")
        print("Constructed")
        subway_frog.__frogs_count += 1 # Accessing class attrbutes using object methods.

    def get_details(self):
        print(f"Name: {self.name}, Color: {self.color}, State: {self.__state}")

    @classmethod
    def get_objs_count(cls): # cls: Starting address of class, self: starting address of object
        print("Total frogs got created: ",cls.__frogs_count)
        
# This main function #
obj1 = subway_frog("red", "joy")
obj1.run()
# print(obj1.__state) #__state: Attribute: Private: Can not access it
print("Color : ", obj1.color) # color: Attribute: Public
#obj1.__get_power() # __get_power: Method: Private
obj1.get_details()


#### 
print("Class attribute")
obj2 = subway_frog()
obj3 = subway_frog()

#print(subway_frog.__frogs_count) # Accessing class attribute
subway_frog.get_objs_count()
```

```python
class Complex:
    def __init__(self, real, im):
        self.real = real
        self.im = im

    def __add__(self, other):
        result = Complex(self.real+other.real, self.im+other.im)
        return result
    def __str__(self):
        return f"Real: {self.real}, Im: {self.im}"
    def __sub__(self, other):
        result = Complex(self.real-other.real, self.im-other.im)
        return result
    def __mul__(self,other):
        result = complex(self.real*other.real,self.im*other.im)
        return result
    def __truediv__(self,other):
        result = complex(self.real/other.real,self.im/other.im)
        return result        
c1 = Complex(1,2)
c2 = Complex(3,4)
print(c1+c2)
print(c1-c2)
print(c1*c2)
print(c1/c2)
```
