```python
import pandas as pd
left_frame = pd.DataFrame({'key': ['Bu','sha','pant'],'match1': [3,4,1],},index=[1,2,3])
right_frame = pd.DataFrame({'key': ['Bu','sha','bhu'],'match1': [3,4,2],},index=[1,2,3])
print(left_frame)
print('\n')
print(right_frame)

```

```python
print(pd.merge(left_frame,right_frame, on='key', how='inner'))#,on=True)
print(pd.merge(left_frame,right_frame, on='key', how='outer'))#,on=True)
```

```python
print(pd.merge(left_frame,right_frame, on='key',how='left'))#,on=True)
print(pd.merge(left_frame,right_frame, on='key', how='right'))#,on=True)
```
